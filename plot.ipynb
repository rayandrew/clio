```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
METRICS = [
    #
    "iops",
    "read_iops",
    "write_iops",
    #
    "write_ratio",
    "read_ratio",
    #
    "iat_avg",
    "read_iat_avg",
    "write_iat_avg",
]


def norm_mean(x: pd.Series):
    return x / x.mean()


def norm_min(x: pd.Series):
    return x / x.min()


def norm_df(df: pd.DataFrame):
    df = df.copy()
    for metric in METRICS:
        df[metric] = norm_mean(df[metric])
        # df[metric] = norm_min(df[metric])
    return df
```
